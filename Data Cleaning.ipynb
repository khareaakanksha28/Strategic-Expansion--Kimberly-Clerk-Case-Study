```python
count = 0
count_3 = 0
incorrect = []
l1 = []
for i, value in enumerate(df['Geographic Area Name']):
    ls = value.split(";")
    count = count + 1
    if len(ls) == 3:
        count_3 = count_3 + 1
    else:
        incorrect.append(i)
        l1.append(value)
    
print(count, count_3)
```

```python
df.loc[14778:15660, 'Geographic Area Name'] = df.loc[14778:15660, 'Geographic Area Name'].str.replace(',', ';')
```

```python
df.loc[14778:15660, 'Geographic Area Name']
```

```python
def county_split(value):
    ls = value.split(';')
    return ls[1]
```

```python
df['County'] = df['Geographic Area Name'].apply(county_split)
df.head()
```

```python
def state_split(value):
    ls = value.split(';')
    return ls[2]
```

```python
df['State'] = df['Geographic Area Name'].apply(state_split)
df.head()
```

```python
df.drop(['Geographic Area Name'], axis = 1, inplace=True)
```

```python
df.head()
```

```python
df['AGE_25_to_29'] = df.groupby(['County', 'State'])['Count!!SEX AND AGE!!Total population!!25 to 29 years'].transform('sum')
df['AGE_30_to_34'] = df.groupby(['County', 'State'])['Count!!SEX AND AGE!!Total population!!30 to 34 years'].transform('sum')
df['AGE_35_to_39'] = df.groupby(['County', 'State'])['Count!!SEX AND AGE!!Total population!!35 to 39 years'].transform('sum')
df['AGE_40_to_44'] = df.groupby(['County', 'State'])['Count!!SEX AND AGE!!Total population!!40 to 44 years'].transform('sum')
df['AGE_45_to_49'] = df.groupby(['County', 'State'])['Count!!SEX AND AGE!!Total population!!45 to 49 years'].transform('sum')
df['AGE_50_to_54'] = df.groupby(['County', 'State'])['Count!!SEX AND AGE!!Total population!!50 to 54 years'].transform('sum')
df['AGE_55_to_59'] = df.groupby(['County', 'State'])['Count!!SEX AND AGE!!Total population!!55 to 59 years'].transform('sum')
df.drop(df.columns[1:20], axis=1, inplace=True)
df.head()
```

```python
df.drop(df.columns[0:5], axis=1, inplace=True)
df.head()
```

```python
df['Male_population'] = df.groupby(['County', 'State'])['Count!!SEX AND AGE!!Male population'].transform('sum')
```

```python
df.head()
```

```python
df.drop(df.columns[0:24], axis=1, inplace=True)
df.head()
```

```python
df['Female_population'] = df.groupby(['County', 'State'])['Count!!SEX AND AGE!!Female population'].transform('sum')
df.head()
```

```python
df.drop(df.columns[0:27], axis=1, inplace=True)
df.head()
```

```python
df['White_population'] = df.groupby(['County', 'State'])['Count!!RACE!!Total population!!One Race!!White'].transform('sum')
df['Black_population'] = df.groupby(['County', 'State'])['Count!!RACE!!Total population!!One Race!!Black or African American'].transform('sum')
df['Asian_population'] = df.groupby(['County', 'State'])['Count!!RACE!!Total population!!One Race!!Asian'].transform('sum')
df.head()
```

```python
df.drop(df.columns[0:16], axis=1, inplace=True)
df.head()
```

```python
df['Hispanic/Latino_population'] = df.groupby(['County', 'State'])['Count!!HISPANIC OR LATINO!!Total population!!Hispanic or Latino (of any race)'].transform('sum')
df['Non-Hispanic/Latino_population'] = df.groupby(['County', 'State'])['Count!!HISPANIC OR LATINO!!Total population!!Not Hispanic or Latino'].transform('sum')
df.head()
```

```python
df.drop(df.columns[0:41], axis=1, inplace=True)
df.head()
```

```python
df['Married_population'] = df.groupby(['County', 'State'])['Count!!HOUSEHOLDS BY TYPE!!Total households!!Married couple household'].transform('sum')
df['Married_population_with_children'] = df.groupby(['County', 'State'])['Count!!HOUSEHOLDS BY TYPE!!Total households!!Married couple household!!With own children under 18 [3]'].transform('sum')
df.head()
```

```python
df['Count!!HOUSEHOLDS BY TYPE!!Total households!!Cohabiting couple household'] = df['Count!!HOUSEHOLDS BY TYPE!!Total households!!Cohabiting couple household'] + df['Count!!HOUSEHOLDS BY TYPE!!Total households!!Male householder, no spouse or partner present:'] + df['Count!!HOUSEHOLDS BY TYPE!!Total households!!Female householder, no spouse or partner present:']
df['Unmarried_population'] = df.groupby(['County', 'State'])['Count!!HOUSEHOLDS BY TYPE!!Total households!!Cohabiting couple household'].transform('sum')
df.head()
```

```python
df.drop(df.columns[0:25], axis=1, inplace=True)
df.head()
```

```python
df['Owned_household_population'] = df.groupby(['County', 'State'])['Count!!HOUSING TENURE!!Occupied housing units!!Owner-occupied housing units'].transform('sum')
df['Rented_household_population'] = df.groupby(['County', 'State'])['Count!!HOUSING TENURE!!Occupied housing units!!Renter-occupied housing units'].transform('sum')
df.head()
```

```python
df.drop(df.columns[0:3], axis=1, inplace=True)
df.head()
```

```python
df_new = df.drop_duplicates()
df_new.head()
```

```python
df_new.columns
```

```python
df_new['Country'] = 'United States'
```

```python
df_new['AGE_Sum'] = df_new['AGE_25_to_29'] + df_new['AGE_30_to_34'] + df_new['AGE_35_to_39'] + df_new['AGE_40_to_44'] + df_new['AGE_45_to_49'] + df_new['AGE_50_to_54'] + df_new['AGE_55_to_59']
```

```python
df_new = df_new.sort_values(by = ['AGE_Sum', 'AGE_25_to_29', 'AGE_30_to_34', 'AGE_35_to_39',
       'AGE_40_to_44', 'AGE_45_to_49', 'AGE_50_to_54', 'AGE_55_to_59',
       'Male_population', 'Female_population', 'White_population',
       'Black_population', 'Asian_population', 'Hispanic/Latino_population',
       'Non-Hispanic/Latino_population', 'Married_population',
       'Married_population_with_children', 'Unmarried_population',
       'Owned_household_population', 'Rented_household_population'], ascending=False)
df_new.head(21)
```
