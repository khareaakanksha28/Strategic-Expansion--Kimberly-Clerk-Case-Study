```python
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt
 
# Load data

df = pd.read_csv("A_data.csv")
 
# Calculate minimum cost for each mill-county pair

df['Min_Cost'] = df[['Trucking cost', 'Rail cost']].min(axis=1)
 
# Create a pivot table for the heatmap

cost_matrix = df.pivot_table(

    index='Mill County', 

    columns='County', 

    values='Min_Cost', 

    aggfunc='min'  # Use min() to handle duplicates (if any)

)
 
# Plot the heatmap

plt.figure(figsize=(20, 15))

sns.heatmap(

    cost_matrix,

    cmap="viridis",  # Color scale for cost values

    linewidths=0.3,

    linecolor="grey",

    annot=False,  # Hide cell values for readability (set to True to see numbers)

    cbar_kws={'label': 'Minimum Transportation Cost ($)'}

)
 
plt.title("Cost Matrix Heatmap: Mills vs. Target Counties", fontsize=16)

plt.xlabel("Target Counties", fontsize=12)

plt.ylabel("Mills (Mill County)", fontsize=12)

plt.xticks(rotation=90)

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
 
# Load data
df = pd.read_csv("A_data1.csv")
 
# Calculate minimum cost for each mill-county pair
df['Min_Cost'] = df[['Trucking cost', 'Rail cost']].min(axis=1)
 
# Create pivot table
cost_matrix = df.pivot_table(
    index='County', 
    columns='Final Supplier', 
    values='Min_Cost', 
    aggfunc='min'
)
 
# Plot heatmap with values
plt.figure(figsize=(25, 15))  # Increase figure size for readability
sns.heatmap(
    cost_matrix,
    cmap="YlGnBu",
    linewidths=0.3,
    linecolor="grey",
    annot=True,  # Show values in cells
    fmt=".2f",   # Format values to 2 decimal places
    annot_kws={'size': 8},  # Adjust font size of annotations
    cbar_kws={'label': 'Minimum Cost ($)'}
)
 
plt.title("Cost Matrix Heatmap: Mills vs. Counties (Values Shown)", fontsize=18)
plt.xlabel("Target Counties", fontsize=14)
plt.ylabel("Mills (Mill County)", fontsize=14)
plt.xticks(rotation=90, fontsize=8)
plt.yticks(fontsize=8)
plt.tight_layout()

plt.savefig("weighted_heatmap.png", dpi=300, bbox_inches="tight")
plt.show()
```
